```r
height_dat <- starwars %>% drop_na(height, species)


# Kernel density plots for height
# grouped by Species
qplot(height, data=height_dat %>% group_by(species) %>% filter(n() >= 3) , geom="density", fill=species, alpha=I(.5), 
   main="Distribution of Species Height", xlab="Height", 
   ylab="Density")
```

```r
mass_dat <- starwars %>% drop_na(mass, species)


# Kernel density plots for height
# grouped by Species
qplot(mass, data=mass_dat %>% group_by(species) %>% filter(n() >= 2) , geom="density", fill=species, alpha=I(.5), 
   main="Distribution of Species Mass", xlab="Mass", 
   ylab="Density")
```
